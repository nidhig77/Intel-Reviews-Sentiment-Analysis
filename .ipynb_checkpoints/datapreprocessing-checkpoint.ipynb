```python
#converting mrp to int
def rupee_to_int(rupee_str):
    # Remove ‚Çπ symbol and commas
    numeric_value = re.sub(r"‚Çπ|,", "", rupee_str)
    # Convert to integer
    return int(numeric_value)
dataset['mrp'] = df['product_mrp'].apply(rupee_to_int)

dataset['mrp']
```

```python
df=df.drop('product_mrp',axis=1)
df.head()
```

```python
#converting to review date to date type and making country as a separate column
df['country'] = df['review_date'].apply(lambda country: re.search(r'in (.+?) on', country).group(1))
df.head()
```

```python
df['country'].unique()
```

```python
from datetime import datetime

df['review_date'] = df['review_date'].apply(lambda text: re.search(r'\d{1,2} [A-Za-z]+ \d{4}', text).group(0))
df['review_date'] = df['review_date'].apply(lambda x: datetime.strptime(x, "%d %B %Y").date())
df.head()
```

```python
df['product_style'].unique()
```

```python
index_list=df[df['product_style']=='Style Name: CoreI5-13400F'].index
for i in index_list:
    df.loc[i,'mrp']=28999
df[df['product_style']=='Style Name: CoreI5-13400F'].head()
```

```python
index_list=df[df['product_style']=='Style Name: CoreI5-13500'].index
for i in index_list:
    df.loc[i,'mrp']=52743
df[df['product_style']=='Style Name: CoreI5-13500'].head()
```

```python
df['product_id']=df['product_name'].apply(lambda s: re.search(r"\d{4,}[A-Z]{0,}", s).group() if re.search(r"\d{4,}[A-Z]{0,}", s) else None)
```

```python
print(df['product_id'].unique())
```

```python
df['product_style'] = df['product_style'].astype(str)
for i in df.index:
    match = re.search(r"\d{4,}[A-Z]{0,}", df.loc[i, 'product_style'])
    if match:
        df.loc[i, 'product_id'] = match.group()
```

```python
df['product_id'].unique()
```
